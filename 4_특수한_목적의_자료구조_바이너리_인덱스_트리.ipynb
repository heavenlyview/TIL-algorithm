# 바이너리 인덱스 트리 (Binary Indexed Tree)
- **2진법 인덱스 구조**를 활용해 구간 합 문제를 효과적으로 해결해 줄 수 있는 자료구조
- == 펜윅 트리(fenwick tree)


- 정수에 따른 2진수 표기

4 정수에 따른 2진수 표기.JPG

- 0이 아닌 마지막 비트를 찾는 방법
  - 특정한 숫자 K의 0이 아닌 마지막 비트를 찾기 위해서 K & -K를 계산한다.

  4 K & -K 계산 결과 예시.JPG

```python
# K & -K 계산 결과 예시

n = 8

for i in range(n + 1):
  print(i, "의 마지막 비트:", (i & -i))
```
